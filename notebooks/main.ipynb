>> ## Seasonal Purchase Behaviour

```python
def is_seasonal_holiday(date):
    if (date.month == 12 and date.day == 25) or (date.month == 1 and date.day == 1):  # Christmas and New Year
        return 'Christmas/New Year'
    elif (date.month == 4 and (date.day >= 15 and date.day <= 21)):  # Easter
        return 'Easter'
    elif date.month == 2 and date.day == 14:
        return 'Valentine\'s Day'
    elif date.month == 5 and date.day == 8:  # May Day (varies)
        return 'May Day'
    elif date.month == 5 and date.weekday() == 6 and (8 <= date.day <= 14):  # Mother's Day
        return 'Mother\'s Day'
    elif date.month == 6 and date.weekday() == 6 and (15 <= date.day <= 21):  # Father's Day
        return 'Father\'s Day'
    elif date.month == 7 and date.day == 4:
        return 'Independence Day'
    elif date.month == 10 and date.day == 31:
        return 'Halloween'
    elif date.month == 11 and date.day == 26:  # Thanksgiving (approx)
        return 'Thanksgiving'
    elif date.month == 12 and date.day == 25:
        return 'Christmas'
    elif date.month == 12 and date.day == 26:
        return 'Boxing Day'
    elif date.month == 12 and date.day == 31:
        return 'New Year\'s Eve'
    
    return 'Normal'

# Apply the function to create a new column
merged_train['SeasonalHoliday'] = merged_train['Date'].apply(is_seasonal_holiday)
```

```python
# Group by seasonal holiday and calculate average sales
seasonal_behavior = merged_train.groupby('SeasonalHoliday')['Sales'].mean().reset_index()
print(seasonal_behavior)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=seasonal_behavior, x='SeasonalHoliday', y='Sales', palette='Blues')
plt.title('Average Sales During Seasonal Holidays')
plt.xlabel('Seasonal Holiday')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.show()
```

>> ## Promo

```python
def check_promo_month(row):
    month2str = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',              
                 7:'Jul', 8:'Aug', 9:'Sept', 10:'Oct', 11:'Nov', 12:'Dec'}
    try:
        months = (row['PromoInterval'] or '').split(',')
        if row['Promo2Open'] and month2str[row['Month']] in months:
            return 1
        else:
            return 0
    except Exception:
        return 0

def promo_cols(df):
    # Months since Promo2 was open
    df['Promo2Open'] = 12 * (df.Year - df.Promo2SinceYear) +  (df.WeekOfYear - df.Promo2SinceWeek)*7/30.5
    df['Promo2Open'] = df['Promo2Open'].map(lambda x: 0 if x < 0 else x).fillna(0) * df['Promo2']
    # Whether a new round of promotions was started in the current month
    df['IsPromo2Month'] = df.apply(check_promo_month, axis=1) * df['Promo2']
```

```python
promo_cols(merged_train)
promo_cols(merged_test)
```
